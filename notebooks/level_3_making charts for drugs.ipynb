```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data

```python
df = pd.read_csv('../notebooks/final_data.csv')
df.head()
```

# The total number of visited patients given the drug molecule along with the categories (up, down, steady, no).

# 1. Metformin 

```python
df['metformin'].value_counts()
```

```python
metformin = df.loc[df['metformin'] != 'No']['metformin'].value_counts()
metformin
```

```python
fig = plt.figure()
ax = fig.add_subplot()

# plotting line

ax.pie(metformin, labels=metformin.index, autopct='%1.1f%%')

# setting title
ax.set_title('The composition of Dosage levels prescribed to the patients')

plt.show()
```
