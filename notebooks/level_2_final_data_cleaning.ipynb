```python
df_new = df_copy.drop(['weight', 'payer_code'], axis=1)
df_new.head()
```

# Changing the admission type values w.r.t its number

```python
df_new['admission_type_id']
```

```python
df_new['admission_type_id'] = df_new['admission_type_id'].map({
    
    1: 'Emergency',
    2: 'Urgent',
    3: 'Elective',
    4: 'Newborn',
    5: 'Not Available',
    6: 'NULL',
    7: 'Trauma Center',
    8: 'Not Mapped'
    
})

df_new.head()
```
